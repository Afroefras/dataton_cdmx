## Transformar

### Unir con resumen

```python
# Unir el resumen a la tabla original

df = df.merge(resumen, on='cluster', how='left').fillna({'nombre':'SIN_CLUSTER'})
df.sample()
```

```python
aux = df.pivot_table(index=['entidad'], columns='nombre', aggfunc={'mes':'count'})
aux['sum'] = aux.sum(axis=1)
for col in aux.columns: aux[col] /= aux['sum']
aux.iloc[:,:-1].fillna(0).style.format("{:.1%}").background_gradient('Blues', axis=1)
```
