# Pronóstico

## Parámetros

```python
BASE_DIR = '/Users/efraflores/Desktop/EF/Contests/Datathon_CDMX/data' # Mac
# BASE_DIR = r'D:\Users\efras\Documents\EF\Contests\Datathon_CDMX\data' # Windows
FILE_NAME = 'ile_limpio_cluster'
```

## Importar

### Clases

```python
from mariachis.ile import InterrupcionEmbarazo
ile = InterrupcionEmbarazo(BASE_DIR, FILE_NAME)
```

### ILE limpio y con cluster

```python
df = ile.full_import(api=False)
df.sample()
```

## Transformar

```python
X, y = ile.build_tad(df, shifts=range(1,22))
X.head(6)
```

## Modelo

### Regresión Lineal

```python
from sklearn.linear_model import LinearRegression

models_linreg, scores_linreg, coef_linreg = ile.train_chunk(X, y, model=LinearRegression, id_cols=['nombre'], to_drop=['fingreso'])
ile.show_scores(scores_linreg)
```

### Red Neuronal

```python
from sklearn.neural_network import MLPRegressor

models_mlp, scores_mlp, coef_mlp = ile.train_chunk(X, y, id_cols=['nombre'], to_drop=['fingreso'], model=MLPRegressor)
ile.show_scores(scores_mlp)
```

## Resultados

### Regresión Lineal

```python
ile.multiplot(X, y, models_linreg, id_cols=['nombre'], from_year=2018, to_year=2020)
```

### Red Neuronal

```python
ile.multiplot(X, y, models_mlp, id_cols=['nombre'], from_year=2018, to_year=2020)
```

## Preservación del modelo

```python
ile.save_model({'pronostico':models_mlp, 'pronostico_cols':X.columns}, 'ile_modelo_pronostico')
```
